# Load libraries

```python
from scipy.special import binom
import matplotlib.pyplot as plt
import matplotlib.ticker as formater

plt.style.use('seaborn-v0_8')
percent_formater = formater.PercentFormatter(xmax=1, decimals=0)
```

# Calculate values

```python
def PWbBiggerThanWa(a, b):
    assert a > b, "a has to be bigger than b"
    return 0.5**a * sum([binom(a + k - 1, k) * 0.5**k for k in range(b)])

b = range(1, 26)
b_plus_1 = [PWbBiggerThanWa(b_val + 1, b_val) for b_val in b]
b_plus_2 = [PWbBiggerThanWa(b_val + 2, b_val) for b_val in b]
b_plus_3 = [PWbBiggerThanWa(b_val + 3, b_val) for b_val in b]
```

# Diagram (English)

```python
_, ax = plt.subplots(figsize=(16, 9))
ax.plot(b, b_plus_1, 'g', label="a=b+1")
ax.plot(b, b_plus_2, 'b', label="a=b+2")
ax.plot(b, b_plus_3, 'r', label="a=b+3")
ax.yaxis.set_major_formatter(percent_formater)
ax.set_xlabel("$b$ (Length of the shorter of the two queues)")
ax.set_ylabel("$P(W_{(b)}>W_{(b+1)})$ (Probability of having to wait longer in the shorter queue)")
ax.legend(loc='lower right')
```
